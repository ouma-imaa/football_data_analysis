```python
# Définir la fonction pour détecter les valeurs aberrantes en utilisant la méthode IQR
def detect_outliers_iqr(df, numeric_vars, threshold=1.5):
    outliers = {}
    for var in numeric_vars:
        # Calculer les quartiles et l'écart interquartile
        Q1 = df[var].quantile(0.25)
        Q3 = df[var].quantile(0.75)
        IQR = Q3 - Q1

        # Identifier les valeurs aberrantes en dehors des seuils supérieurs et inférieurs
        lower_bound = Q1 - threshold * IQR
        upper_bound = Q3 + threshold * IQR
        outliers[var] = df[(df[var] < lower_bound) | (df[var] > upper_bound)][var]
    return outliers

# Utiliser la fonction pour détecter les valeurs aberrantes dans vos données
outliers_iqr = detect_outliers_iqr(df, numeric_vars)

# Afficher les valeurs aberrantes détectées
for var, outliers_list in outliers_iqr.items():
    print(f"Variable : {var}")
    print(outliers_list)
    print("-" * 50)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Déterminer la taille de la grille en fonction du nombre de variables numériques
num_plots = len(numeric_vars)
num_rows = (num_plots + 2) // 3
num_cols = min(num_plots, 3)

# Créer une grille de sous-graphiques pour chaque variable numérique
plt.figure(figsize=(12, 4 * num_rows))
for i, var in enumerate(numeric_vars):
    plt.subplot(num_rows, num_cols, i+1)
    sns.boxplot(y=df[var])
    sns.stripplot(y=outliers_iqr[var], color='red', alpha=0.5)
    plt.title(var)
    plt.tight_layout()

plt.show()
```
